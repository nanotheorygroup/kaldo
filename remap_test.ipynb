```python
from kaldo.forceconstants import ForceConstants
import os
```

### Silicon (216 Atom Supercell, 2 Atom Primitive Cell)

```python
tdep_dir = os.path.abspath("./kaldo/tests/si-tdep2")
# 50 seconds on Ethans Desktop :(
fc = ForceConstants.from_folder(
        folder=str(tdep_dir), format="tdep", supercell = (3,3,3), include_fourth=True,
    )
```

```python
# 1.2 seconds on Ethans Desktop
fc2_sc = fc.irred_to_full(order = 2)
print(fc2_sc.shape)
print(fc2_sc[0,0,:,0,0,:].reshape(3,3))
```

```python
# 3.9 seconds on Ethans Desktop
fc3_sc = fc.irred_to_full(order = 3)
print(fc3_sc.shape)
print(fc3_sc[0,0,:,0,0,:,0,0,:].reshape(3,3,3))
```

```python
fc4_sc = fc.irred_to_full(order = 4)
print(fc4_sc.shape)
print(fc4_sc[0,0,:,0,0,:,0,0,:,0,0,:].reshape(3,3,3,3))
```
